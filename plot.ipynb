```python
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = [16, 8]
matplotlib.rc('font', size=20)
matplotlib.rc('font', family='serif')
matplotlib.rc('lines', linewidth=2.5,linestyle="-.")
matplotlib.rc('lines', markersize=10)
matplotlib.rc('figure.subplot', hspace=.4)

import gaussian

# create a data kind plotting file parameters
```

```python
data = np.loadtxt("parameters.txt")
g_list = [gaussian.Gaussian(row[0], row[1], 1j*row[3], 1, row[4] + 1j*row[5]) for row in data ]
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2)
fig.suptitle('Horizontally stacked subplots')

x = np.linspace(-4,4,10**3)
y = np.zeros(len(x), dtype='complex128')
for G in g_list:
    y += G.psi(x)
    ax1.plot(x, G.psi(x).real)

true = gaussian.Gaussian(np.zeros(1),np.zeros(1),np.ones(1)*1j,1,1)
print(true.l2norm(x))
#ax2.plot(x, y.real, label = 'reconstruction')
#ax2.plot(x, true.real, label = 'original')
ax2.plot(x, np.abs(y - true.psi(x)).real, label = 'abs error.re')
plt.legend()
# we are comparing the integration rules for the momentum variable
```
